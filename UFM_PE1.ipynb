```python
import matplotlib.pyplot as plt

def plot_supply_demand_shift_with_credit_spanish_custom(scenario_number, supply_shift=None, demand_shift=None):
    fig, ax = plt.subplots(figsize=(8, 6))
    ax.set_title(f"Ítem {scenario_number}", fontsize=16, fontweight='bold', color='#333333')
    ax.set_xlabel("Cantidad", fontsize=12, fontweight='bold')
    ax.set_ylabel("Precio", fontsize=12, fontweight='bold')

    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.spines['left'].set_linewidth(1.5)
    ax.spines['bottom'].set_linewidth(1.5)

    ax.tick_params(axis='x', colors='#333333', labelsize=10)
    ax.tick_params(axis='y', colors='#333333', labelsize=10)

    fig.patch.set_facecolor('#f0f0f0')
    ax.set_facecolor('#ffffff')

    quantity = [0, 10]
    original_demand = [10, 0]
    original_supply = [0, 10]

    ax.plot(quantity, original_demand, label='Demanda Original', linestyle='-', color='#1f77b4', linewidth=2)
    ax.plot(quantity, original_supply, label='Oferta Original', linestyle='-', color='red', linewidth=2)

    if supply_shift == 'right':
        shifted_supply = [-2, 8]
        ax.plot(quantity, shifted_supply, label='Nueva Oferta', color='red', linestyle='--', linewidth=2)
    elif supply_shift == 'left':
        shifted_supply = [2, 12]
        ax.plot(quantity, shifted_supply, label='Nueva Oferta', color='red', linestyle='--', linewidth=2)

    if demand_shift == 'right':
        shifted_demand = [12, 2]
        ax.plot(quantity, shifted_demand, label='Nueva Demanda', color='#1f77b4', linestyle='--', linewidth=2)
    elif demand_shift == 'left':
        shifted_demand = [8, -2]
        ax.plot(quantity, shifted_demand, label='Nueva Demanda', color='#1f77b4', linestyle='--', linewidth=2)

    ax.legend(frameon=False, fontsize=10, loc='best')
    plt.grid(True, linestyle='--', color='#cccccc')

    plt.text(7, -5, 'Autor Kevin Lajpop', fontsize=12, color='#555555', ha='center', va='center', fontstyle='italic')

    return fig

fig1_with_credit_spanish_custom = plot_supply_demand_shift_with_credit_spanish_custom(1, supply_shift='right')
fig2_with_credit_spanish_custom = plot_supply_demand_shift_with_credit_spanish_custom(2, demand_shift='right')
fig3_with_credit_spanish_custom = plot_supply_demand_shift_with_credit_spanish_custom(3, supply_shift='right')
fig4_with_credit_spanish_custom = plot_supply_demand_shift_with_credit_spanish_custom(4, supply_shift='left')
fig5_with_credit_spanish_custom = plot_supply_demand_shift_with_credit_spanish_custom(5, demand_shift='right')
fig6_with_credit_spanish_custom = plot_supply_demand_shift_with_credit_spanish_custom(6, demand_shift='left')

fig1_with_credit_spanish_custom.show()
fig2_with_credit_spanish_custom.show()
fig3_with_credit_spanish_custom.show()
fig4_with_credit_spanish_custom.show()
fig5_with_credit_spanish_custom.show()
fig6_with_credit_spanish_custom.show()
```
